```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Step 1: Create synthetic data (e.g., y = 2x + 1 with noise)
X = np.array([[1], [2], [3], [4], [5]])  # Features (2D)
y = np.array([3, 5, 7, 9, 11])           # Labels (1D)

# Step 2: Create and train the model
model = LinearRegression()
model.fit(X, y)

# Step 3: Print the learned parameters
print("Slope (coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)

# Step 4: Make predictions
X_test = np.array([[6], [7]])
y_pred = model.predict(X_test)
print("Predictions for 6 and 7:", y_pred)

# Step 5: Plot the results
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, model.predict(X), color='red', label='Regression line')
plt.scatter(X_test, y_pred, color='green', label='Predictions')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```
